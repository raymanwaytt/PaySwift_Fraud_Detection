# FEATURE ENGINEERING <br>
### Adding time_of_day column : `Morning`, `Afternoon`, `Night`

```python
# Define conditions for each time of the day
condition_night = (fraud_model_data['hour'] >= 20) | (fraud_model_data['hour'] < 6)  # Night is from 8 PM to 6 AM
condition_morning = (fraud_model_data['hour'] >= 6) & (fraud_model_data['hour'] < 12)  # Morning is from 6 AM to 12 PM
condition_afternoon = (fraud_model_data['hour'] >= 12) & (fraud_model_data['hour'] < 20)  # Afternoon is from 12 PM to 8 PM

# Combine the conditions into a list
conditions = [condition_night, condition_morning, condition_afternoon]

# Values corresponding to each condition
values = ['Night', 'Morning', 'Afternoon']

fraud_model_data['time_of_day'] = np.select(conditions, values, default='Unknown')
fraud_model_data.head()
```

# MODELING 

### Spliting the data into `features` and `label`

```python
X = fraud_model_data.drop(columns=['transaction_id', 'is_fraud'])  # dropping redundant and label columns
y = fraud_model_data['is_fraud'].astype('int') # convert to int
X, y
```

### Encoding the Categorical Value Columns i.e *convert the categorical columns to binary using `pd.dummies()`*

```python
# using pd.get_dummies to One-hot-encode categorical values columns
X = pd.get_dummies(X, columns= ['region', 'type', 'time_of_day'], dtype = 'int') # convert to binary classification
X.head()
```

now that the feature and label data are ready lets proceed to :
###  Splitting into `trainig` and `test` data

```python
# splitting data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

##  Training the model using `RandomForestClassifier`

```python
# instantiating the RandomForest estimator
model = RandomForestClassifier(random_state=42)

# training the model
model.fit(X_train, y_train)
```

```python
# checking accuracy of the model on our test data set
model.score(X_test, y_test)
```

#### Even with an high Accuracy Score, since the class is unbalanced this doesn't guarantee the model good enough to detect fraud transactions.

## EVALUATING THE MODEL...

```python
# model evaluation
y_pred = model.predict(X_test)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

### key observation
- The model does well at identifying 90% of legit txn, with a precision of about 95%
- Recall is low for Fraud txn, i.e the model is missing lot of Fraud Cases (TP), not strict on fraud cases-- High FN
- The precision for transaction's flagged is about 0.45, meaning the model is weak & uncertain with it's prediction.

### Tunning the hyperparameters...
since the Class distribution for fraud vs legit cases is inbalanced I'll be increasing the class balance :
- Class 0 (non-fraud): normal weight
- Class 1 (fraud): 10x the weight/importance

#### Now the model pays more attention to fraud cases.
- I'll be setting the threshold to a range of values (0.1 - 0.5), to observe which performs well the most.

```python
# instatiating the model
model = RandomForestClassifier(random_state=42, class_weight={0:1, 1:10}) # tunning the class weight in favor of fraudlent txn

# fitting the model
model.fit(X_train, y_train)

# make predictions with probability
y_proba = model.predict_proba(X_test)[:, 1]  # probability for the POSITIVE(FRAUD} prediction

# setting a threshold between 0.1-0.55 at which the prob of a fraud guess is considered TRUE
thresholds = np.arange(0.1, 0.6, 0.05)

for t in thresholds:
    y_pred = (y_proba >= t).astype(int)
    print(f"""
    Threshold {t:.2f}: \n Classification Report: \n {classification_report(y_test, y_pred)}
    Confusion Matrix: \n {confusion_matrix(y_test, y_pred)}
    """)
```

the lower the threshold, the model gets:

- More aggressive at catching frauds (higher recall)

- But it also makes more false alarms (lower precision)

## Tradeff Plot

```python
def tradeoff_plot(y_proba, thresholds, best_thresh=None):
    """
    This plots a preision-recall-f1_score chart across the provided list of thresholds
    
    This func takes in 3 arg, 
    - y_proba : the probability of a certain class (in this case class 1 (fraud cases))
    - thresholds : A list of thresholds values for the model to test through
    - best_thresh (default= None): A single value representing the best threshold value, this plots a line to show the best threshold value and where it cuts accross other metrics
    """

    # imports all libraries
    from sklearn.metrics import precision_score, recall_score, f1_score
    import matplotlib.pyplot as plt

    # list of all metrics
    precision_list = []
    recall_list = []
    f1_list = []

    # loops through the thresholds values 
    for t in thresholds:
        # if the prob value > threshold sets to 1, otherwise 0
        y_pred = (y_proba >= t).astype(int)
        
        # Evaluates the pred value and appends each metrics score to the metrics list
        precision_list.append(precision_score(y_test, y_pred))
        recall_list.append(recall_score(y_test, y_pred))
        f1_list.append(f1_score(y_test, y_pred))
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(thresholds, precision_list, label='Precision', marker='o')
    plt.plot(thresholds, recall_list, label='Recall', marker='o')
    plt.plot(thresholds, f1_list, label='F1 Score', marker='o')

    # if best threshold is provided, plots a line at that spot to cut accross the graph
    if best_thresh is not None:
        plt.axvline(x=best_thresh, color='red', linestyle='--', label=f'Selected Threshold: {best_thresh}')
    
    plt.xlabel('Threshold')
    plt.ylabel('Score')
    plt.title('Precision, Recall, and F1 vs Threshold')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

### plotting trade-off for RF Classifier...

```python
thresholds = np.arange(0.1, 1, 0.05)
y_proba = model.predict_proba(X_test)[:, 1]

# plots trade-off graph with a best threshold value of 0.2
tradeoff_plot(y_proba, thresholds, .2)
```

### Threshold Tuning Insights
#### This plot shows how precision, recall, and F1-score vary as we adjust the classification threshold.

- As expected, recall decreases as the threshold increases — the model gets stricter, so it catches fewer positives.

- Precision increases with the threshold — we get fewer false positives, but at the cost of missing actual frauds.

- The F1-score (which balances both) peaks around `threshold = 0.20`, making it our optimal sweet spot.

#### The red dashed line marks this chosen threshold. It offers a good compromise between catching fraudulent transactions and avoiding too many false alarms.

## Plotting the Receiver Operating Characteristic (ROC) Curve

### Trying other approach to get better results...

## Applying `SMOTE` (Synthetic Minority Oversampling Technique)
SMOTE creates synthetic fraud cases based on existing ones, so the model sees more fraud cases

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to balance the training set
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_train_resampled.value_counts())
```

Now the cases of fraud is equal to that of legit cases, 
#### fitting the RF model once again...

```python
# Train model
model = RandomForestClassifier(random_state=42)
model.fit(X_train_resampled, y_train_resampled)

# Predict
y_proba = model.predict_proba(X_test)[:, 1]

# setting a threshold between 0.1-0.55 at which the prob of a fraud guess is considered TRUE
thresholds = np.arange(0.1, 0.6, 0.05)

for t in thresholds:
    y_pred = (y_proba >= t).astype(int)
    print(f"""
    Threshold {t:.2f}: \n Classification Report: \n {classification_report(y_test, y_pred)}
    Confusion Matrix: \n {confusion_matrix(y_test, y_pred)}
    """)

# # Evaluate
# print("Classification Report:\n", classification_report(y_test, y_pred))
# print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# plotting precision recall graph

thresholds = np.arange(0.1, 0.6, 0.05)
y_proba = model.predict_proba(X_test)[:, 1]

# selected threshold
tradeoff_plot(y_proba, thresholds, .2)
```

### Observations:
- Without SMOTE, recall is low (0.25), missing many frauds.
- With SMOTE and threshold=0.2, recall improves significantly (0.86), but precision remains approx 0.4, leading to many false positives (345 FP vs. 216 TP).
- Best F1-score (0.53 at threshold=0.2) shows a trade-off favoring recall over precision.

### Precision-Recall-Trade-Off plot for LogisticRegression

### AUC Score `Logistics Regression`

```python
roc_auc = roc_auc_score(y_test, y_proba)
print("AUC Score:", roc_auc)
```

### Observations:

- At threshold=0.5, recall drops drastically (0.4), missing most frauds, though precision improves slightly (0.40).
- F1-score peaks at 0.47 (threshold=0.2), indicating a worse balance than Random Forest.

## trying `XGBoost Classifier` algorithm to see if it learns better

#### Plotting a trade-off chart for `XGBoost Classifier`

```python
thresholds = np.arange(0.1, 1, 0.05)
y_pred_xgb = xgb_model.predict_proba(X_test)[:, 1]

# selected threshold
tradeoff_plot(y_pred_xgb, thresholds, .2)
```

### AUC/ROC - `XGBoost`

```python
y_proba_xgb = xgb_model.predict_proba(X_test)[:, 1] # Probabilities for class 1

fpr, tpr, threshold = roc_curve(y_test, y_proba_xgb)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.grid()
plt.show()
```

- Highest Recall for Fraud Detection: Random Forest (SMOTE) achieves the highest recall of 0.86 at threshold 0.20, catching 216 out of 252 frauds in the test set. This maximizes financial savings (N8.208M at N38K per fraud), critical for minimizing losses in a high-stakes fintech context. XGBoost (0.75 recall, 188 frauds) and Logistic Regression (0.77 recall, 193 frauds) catch fewer frauds, saving N7.144M and N7.334M, respectively.
- Best F1-Score for Balance: With an F1-score of 0.53, Random Forest (SMOTE) offers the best balance between recall and precision, tied with Random Forest (class-balanced) but with higher recall. XGBoost’s F1-score (0.51) is slightly lower, and Logistic Regression’s (0.47) is the weakest, reflecting poor precision (0.34).
- Manageable False Positives: Random Forest (SMOTE) has 345 false positives, higher than XGBoost’s 299 but lower than Logistic Regression’s 380. While this increases operational costs (N1.725M at N5K per FP), the additional frauds caught (216 vs. 188 for XGBoost) save an extra N1.064M, resulting in the highest net savings (N6.483M vs. N5.649M for XGBoost, N5.434M for LR).
- Robust to Imbalance: Using SMOTE, Random Forest effectively handles the imbalanced dataset (13% fraud), outperforming Logistic Regression, which struggles with low precision and high FPs. XGBoost also benefits from SMOTE but falls short on recall compared to Random Forest.
- Reliable Performance: Random Forest (SMOTE) likely matches XGBoost’s AUC-ROC of 0.88 (as seen with RF class-balanced at 0.882), indicating strong discrimination ability, while Logistic Regression lags (AUC ~0.75-0.80).
